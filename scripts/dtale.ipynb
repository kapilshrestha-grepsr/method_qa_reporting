```python
import pandas as pd
import dtale
import os
```

```python
file_name = "method_summary_report.csv"
```

```python
# List your file paths here
file_paths = [
"/home/kapil/project/method_qa_reporting/methodusa_parent/qa_reports/method_summary_report.csv"
]

# Dictionary to hold dataframes and D-Tale instances
dtale_instances = {}

# Load and launch D-Tale for each file
for idx, path in enumerate(file_paths, start=1):
    try:
        df = pd.read_csv(path, dtype=str)
        df_name = f"df{idx}"
        print(f"{df_name} loaded with shape: {df.shape}")
        instance = dtale.show(df, name=df_name, dtype=str)
        instance.open_browser()
        dtale_instances[df_name] = instance
    except Exception as e:
        print(f"Failed to load {path}: {e}")
```
